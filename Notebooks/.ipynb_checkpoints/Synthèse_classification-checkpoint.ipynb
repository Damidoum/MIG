# Synthèse classification

Dans ce notebook nous présentons les résulats obtenus en faisant varier le nombre de classe. 

## Données générales

```python
#importation des librairies 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
#from tslearn.clustering import TimeSeriesKMeans
import seaborn as sns
```

```python
timeSeries = pd.read_csv('../data/timeSeries_epsilon.csv', index_col = 'node')
```

```python
timeSeries
```

```python
node = [int(x.split('_')[1]) for x in list(timeSeries.index)]
```

```python
nodes = {}
for x in timeSeries.index : 
    nodes[x] = int(x.split('_')[1])
coord = pd.read_csv('../data/coord.csv')
coord.set_index('node', inplace = True)
```

## Pour un clustering à 50 classes 

```python
with open("../data/Classifcation/cluster_50.txt", 'r') as f:
     classification_50 = [int(line.rstrip('\n')) for line in f]
```

```python
df_classification_50 = pd.DataFrame(np.array([node, classification_50]).transpose(), columns = ['node', 'classe'])
eps = lambda x : 'epsilon_' + str(x)
df_classification_50['node'] = df_classification_50['node'].apply(eps)
df_classification_50.set_index('node', inplace = True)
df_classification_50 = pd.concat([df_classification_50, timeSeries],axis =1)
```

```python
# Affichage de toute la palette : 
fig, axes = plt.subplots(10,5,figsize = (20,20), sharex = True, sharey = True)

for k in range(len(classification_50)): 
    axes[classification_50[k]//5, classification_50[k]%5].plot(timeSeries.iloc[k,:])
```

```python
#Affichage d'un représentant par classe : 
df_group_50 = df_classification_50.groupby('classe')
df_represent_50 = df_group_50.first().sort_index()
```

```python
df_represent_50.transpose().plot(figsize = (20,20));
```

### Poids de chaque classe

```python
df_group_50.size().plot.bar(figsize= (10, 5));
```

### Cartographie pour 50 classes

```python
carte_50 = pd.concat([coord, df_classification_50.rename(nodes)['classe']], axis = 1)
carte_50 = carte_50.drop('Y', axis = 1)
```

```python
sns.lmplot( x="X", y="Z", data=carte_50, fit_reg=False, hue='classe', legend = False, palette = 'Spectral');
```
