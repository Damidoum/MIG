```python
import pandas as pd
import numpy as np
```

```python
sdf=pd.read_csv('timeSeries_classe_10.csv', index_col='node')
sdf
```

```python
def lhs_method(sdf):
    LHS=[]
    sampler = qmc.LatinHypercube (d=2)
    for group in sdf.groupby('classe'):
        plage=group[1].describe().loc[['min','max']][['300.0','3630.0']]
        sample = sampler.random (n=10)
        sample[:,0]=sample[:,0]*plage.loc['min']['300.0']+(1-sample[:,0])*plage.loc['max']['300.0']
        sample[:,1]=sample[:,1]*plage.loc['min']['3630.0']+(1-sample[:,1])*plage.loc['max']['3630.0']
        LHS.append(sample)
    return LHS
LHS=lhs_method(sdf)
LHS
```
